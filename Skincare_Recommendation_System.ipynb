```python
import pandas as pd

data = {
    'product_id': [1, 2, 3, 4, 5],
    'product_name': [
        'SKIN1004 Madagascar Centella Ampoule',
        'Anua Heartleaf 77% Soothing Toner',
        'SKIN1004 Centella Light Cleansing Oil',
        'Anua Heartleaf Pore Control Cleansing Oil',
        'The Ordinary Hyaluronic Acid 2% + B5'
    ],
    'brand': ['SKIN1004', 'Anua', 'SKIN1004', 'Anua', 'The Ordinary'],
    'category': ['Ampoule', 'Toner', 'Cleansing Oil', 'Cleansing Oil', 'Serum'],
    'skin_type': ['Oily Dry Sensitive Combination', 'Oily Sensitive Combination', 'Oily Combination', 'Oily Sensitive Combination', 'Dry Sensitive'],
    'concerns': ['Acne Redness Irritation', 'Redness Irritation Acne', 'Blackheads Sebum', 'Blackheads Acne LargePores', 'Dehydration Dryness']
}

df = pd.DataFrame(data)
print("--- قاعدة بيانات المنتجات جاهزة ---")
display(df)
```

```python
def recommend_skincare(user_skin_type, user_concern):
    matched_stage1 = df[df['skin_type'].str.contains(user_skin_type, case=False)]

    def calculate_score(row):
        product_concerns = row['concerns'].split()
        score = sum(1 for concern in product_concerns if concern.lower() in user_concern.lower())
        return score

    matched_stage2 = matched_stage1.copy()
    matched_stage2['matching_score'] = matched_stage2.apply(calculate_score, axis=1)

    recommendations = matched_stage2.sort_values(by='matching_score', ascending=False)

    return recommendations[['product_name', 'brand', 'category', 'matching_score']]

print("--- دالة التوصية الذكية تم بناؤها بنجاح ---")
```

```python
my_skin_type = 'Sensitive'
my_concerns = 'Acne Redness LargePores'

my_recommendations = recommend_skincare(my_skin_type, my_concerns)

print(f"--- التوصيات الذكية المخصصة للبشرة ({my_skin_type}) ---")
display(my_recommendations)
```
